```python
import pandas as pd
import matplotlib.pyplot as plt

# Crear un DataFrame con datos históricos de exportaciones e importaciones
data = {
    'Año': [2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025],
    'Exportaciones (EE.UU.)': [120000, 130000, 115000, 120000, 125000, 130000, 132000, 135000],  # En millones USD
    'Importaciones (China)': [450000, 460000, 420000, 430000, 440000, 445000, 450000, 460000],  # En millones USD
}

df = pd.DataFrame(data)

# Calcular la balanza comercial (Exportaciones - Importaciones)
df['Balanza Comercial'] = df['Exportaciones (EE.UU.)'] - df['Importaciones (China)']

# Mostrar los datos
print(df)

```

```python
# --- Gráfica 1: Evolución de Exportaciones e Importaciones ---
plt.figure(figsize=(10, 6))
plt.plot(df['Año'], df['Exportaciones (EE.UU.)'], marker='o', label='Exportaciones (EE.UU.)')
plt.plot(df['Año'], df['Importaciones (China)'], marker='o', label='Importaciones (China)')
plt.title('Evolución de Exportaciones e Importaciones EE.UU. - China (2018-2025)')
plt.xlabel('Año')
plt.ylabel('Valor en millones USD')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Gráfica 2: Balanza Comercial (barras con color según signo) ---
plt.figure(figsize=(10, 6))
colors = ['red' if x < 0 else 'green' for x in df['Balanza Comercial']]
plt.bar(df['Año'].astype(str), df['Balanza Comercial'], color=colors, alpha=0.7)
plt.title('Balanza Comercial EE.UU. - China (2018-2025)')
plt.xlabel('Año')
plt.ylabel('Balanza Comercial en millones USD')
plt.axhline(0, color='black', linewidth=0.8)  # línea en 0
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Convertir los datos a formato de regresión
X = np.array(df['Año']).reshape(-1, 1)  # Variable independiente (Año)
y_exportaciones = np.array(df['Exportaciones (EE.UU.)']).reshape(-1, 1)  # Exportaciones
y_importaciones = np.array(df['Importaciones (China)']).reshape(-1, 1)   # Importaciones

# Crear y entrenar los modelos de regresión
model_exportaciones = LinearRegression()
model_importaciones = LinearRegression()

model_exportaciones.fit(X, y_exportaciones)
model_importaciones.fit(X, y_importaciones)

# Predicción para el futuro
pred_exportaciones = model_exportaciones.predict(X)
pred_importaciones = model_importaciones.predict(X)

# Graficar los resultados
plt.figure(figsize=(10, 6))
plt.plot(df['Año'], y_exportaciones, marker='o', label='Exportaciones (EE.UU.)')
plt.plot(df['Año'], pred_exportaciones, linestyle='--', label='Tendencia Exportaciones')
plt.plot(df['Año'], y_importaciones, marker='o', label='Importaciones (China)')
plt.plot(df['Año'], pred_importaciones, linestyle='--', label='Tendencia Importaciones')
plt.title('Tendencia de Exportaciones e Importaciones EE.UU.-China (2018-2025)')
plt.xlabel('Año')
plt.ylabel('Valor en millones USD')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos reales del comercio EE.UU.-China 2025 (ene-abr)
data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril'],
    'Exportaciones': [9991.3, 10461.6, 11458.2, 8193.1],
    'Importaciones': [41639.2, 31635.4, 29383.7, 25378.1]
}

df = pd.DataFrame(data)

df['Balanza'] = df['Exportaciones'] - df['Importaciones']

# Gráfico de exportaciones e importaciones
plt.figure(figsize=(10, 6))
plt.plot(df['Mes'], df['Exportaciones'], marker='o', label='Exportaciones')
plt.plot(df['Mes'], df['Importaciones'], marker='o', label='Importaciones')
plt.title('Comercio bilateral EE.UU.-China (enero-abril 2025)')
plt.xlabel('Mes')
plt.ylabel('Millones de USD')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Gráfico de balanza comercial
plt.figure(figsize=(10, 6))
plt.bar(df['Mes'], df['Balanza'], color=['red' if x < 0 else 'green' for x in df['Balanza']])
plt.title('Balanza Comercial EE.UU.-China (enero-abril 2025)')
plt.xlabel('Mes')
plt.ylabel('Millones de USD')
plt.axhline(0, color='black', linewidth=0.8)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos por sectores (ene-abr 2025)
data = {
    'Sector': [
        'Electrónica', 'Semiconductores', 'Televisores',
        'Agricultura', 'Maquinaria', 'Aeronaves'
    ],
    'Exportaciones_Ene': [3200, 1500, 700, 13000, 9500, 6700],
    'Exportaciones_Abr': [2500, 1100, 500, 9800, 7400, 5100],
    'Importaciones_Ene': [27000, 18000, 10000, 4300, 2200, 1100],
    'Importaciones_Abr': [20000, 12500, 7000, 2800, 1500, 800]
}

df = pd.DataFrame(data)

# Calcular variaciones porcentuales
df['Var_Exp_%'] = ((df['Exportaciones_Abr'] - df['Exportaciones_Ene']) / df['Exportaciones_Ene']) * 100
df['Var_Imp_%'] = ((df['Importaciones_Abr'] - df['Importaciones_Ene']) / df['Importaciones_Ene']) * 100

# Visualización de exportaciones
plt.figure(figsize=(12, 6))
bar_width = 0.35
index = range(len(df))

plt.bar(index, df['Exportaciones_Ene'], bar_width, label='Exportaciones Enero')
plt.bar([i + bar_width for i in index], df['Exportaciones_Abr'], bar_width, label='Exportaciones Abril')
plt.xticks([i + bar_width / 2 for i in index], df['Sector'])
plt.ylabel('Millones USD')
plt.title('Comparación de Exportaciones por Sector (Ene vs Abr 2025)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Visualización de importaciones
plt.figure(figsize=(12, 6))
plt.bar(index, df['Importaciones_Ene'], bar_width, label='Importaciones Enero')
plt.bar([i + bar_width for i in index], df['Importaciones_Abr'], bar_width, label='Importaciones Abril')
plt.xticks([i + bar_width / 2 for i in index], df['Sector'])
plt.ylabel('Millones USD')
plt.title('Comparación de Importaciones por Sector (Ene vs Abr 2025)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Mostrar tabla con variaciones
print(df[['Sector', 'Var_Exp_%', 'Var_Imp_%']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt

sectores = ['Smartphones', 'Ordenadores', 'Automóviles', 'Televisores', 'Maquinaria']
coste_base = [200, 400, 15000, 300, 12000]  # Coste promedio por unidad sin aranceles
proporcion_china = [0.4, 0.35, 0.2, 0.5, 0.25]  # Proporción del coste vinculado a componentes chinos

# Cálculo del coste adicional por arancel
coste_adicional = [base * prop * 0.25 for base, prop in zip(coste_base, proporcion_china)]
coste_total = [base + extra for base, extra in zip(coste_base, coste_adicional)]

df = pd.DataFrame({
    'Sector': sectores,
    'Coste_base': coste_base,
    'Coste_adicional_arancel': coste_adicional,
    'Coste_total_estimado': coste_total
})

# Visualización
plt.figure(figsize=(10, 6))
plt.bar(df['Sector'], df['Coste_base'], label='Coste base', alpha=0.7)
plt.bar(df['Sector'], df['Coste_adicional_arancel'], bottom=df['Coste_base'], label='Arancel 25%', alpha=0.9)
plt.ylabel('Coste por unidad (USD)')
plt.title('Impacto estimado de aranceles sobre sectores tecnológicos')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(df) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Nuevos datos: estimación de aumento del coste de producción en sectores afectados
sectores = ['Electrónica de consumo', 'Automoción', 'Farmacéutica', 'Maquinaria', 'Telecomunicaciones']
coste_2024 = [100, 15000, 500, 12000, 8000]  # Coste promedio por unidad en 2024
arancel_china = [0.3, 0.2, 0.15, 0.25, 0.35]  # Proporción del coste expuesto a componentes chinos

# Simulación de arancel del 25%
aumento_arancel = [base * prop * 0.25 for base, prop in zip(coste_2024, arancel_china)]
coste_2025 = [base + extra for base, extra in zip(coste_2024, aumento_arancel)]

df_costes = pd.DataFrame({
    'Sector': sectores,
    'Coste_2024': coste_2024,
    'Aumento_por_arancel': aumento_arancel,
    'Coste_2025_estimado': coste_2025
})

# Visualización
plt.figure(figsize=(10, 6))
bar_width = 0.5
index = range(len(df_costes))

plt.bar(index, df_costes['Coste_2024'], bar_width, label='Coste 2024', alpha=0.7)
plt.bar(index, df_costes['Aumento_por_arancel'], bar_width, bottom=df_costes['Coste_2024'],
        label='Aumento por Arancel 2025', alpha=0.9)

plt.xticks(index, df_costes['Sector'], rotation=45)
plt.ylabel('Coste por unidad (USD)')
plt.title('Simulación del impacto arancelario en sectores industriales')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(df_costes) 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Datos simulados de diversificación geográfica de la producción (porcentaje)
data = {
    'Año': [2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025],
    'China': [80, 78, 75, 72, 68, 63, 60, 55],
    'India': [5, 6, 7, 8, 10, 13, 15, 17],
    'Vietnam': [3, 4, 5, 6, 8, 10, 12, 13],
    'Otros': [12, 12, 13, 14, 14, 14, 13, 15]
}

df_diversificacion = pd.DataFrame(data)

# Conversión a arrays tipo float
x = np.array(df_diversificacion['Año'], dtype=float)
china = np.array(df_diversificacion['China'], dtype=float)
india = np.array(df_diversificacion['India'], dtype=float)
vietnam = np.array(df_diversificacion['Vietnam'], dtype=float)
otros = np.array(df_diversificacion['Otros'], dtype=float)

# Gráfico de líneas apiladas
plt.figure(figsize=(10, 6))
plt.stackplot(x, china, india, vietnam, otros, labels=['China', 'India', 'Vietnam', 'Otros'])
plt.title('Evolución de la Diversificación Geográfica de la Producción Tecnológica')
plt.xlabel('Año')
plt.ylabel('Porcentaje (%)')
plt.legend(loc='upper right')
plt.grid(True)
plt.tight_layout()
plt.show()

print(df_diversificacion)
```
